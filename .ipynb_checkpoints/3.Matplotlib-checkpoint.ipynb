```python
import numpy as np
import matplotlib.pyplot as plt
```

- простые графики

```python
# симулируем данные
# 1-мерный массив целых значений
x = np.arange(100)
x
```

```python
# рисуем график
y = x.copy()

plt.plot(x, y)
plt.show()
```

```python
# 1-мерный массив из 1000 значений
# из линейной последовательности в диапазоне от 0 до 10
x = np.linspace(0, 10, 1000)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
# совмещение нескольких графиков в одном окне
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
# Настройка графика: цвета и вид линий
# цвета можно задавать разными способами
plt.plot(x, np.sin(x), color="green") # по названию
plt.plot(x, np.sin(x+1), color="k") # буква из rgbcmyk
plt.plot(x, np.sin(x+2), color=(1.0, 0.2, 0.0)) # кортеж со значениями от 0 до 1 в формате RGB
plt.plot(x, np.sin(x+3), color="#ff11dd") # 16-ричный код со значениями от 00 до ff в формате RGB
plt.show()
```

```python
# стиль линий
plt.plot(x, np.sin(x), color="green", linestyle="solid") # сплошная лния
plt.plot(x, np.sin(x-1), color="r", linestyle="dashed") # штриховая лния
plt.plot(x, np.sin(x-2), color="b", linestyle="dotted") # пунктирная лния
plt.plot(x, np.sin(x-3), color="y", linestyle="dashdot") # штрихпунктирная лния
plt.show()
```

```python
# пределы осей
plt.plot(x, np.sin(x), color="green", linestyle="solid")
plt.xlim(-5, 15)
plt.ylim(-2, 2)
plt.show()
```

```python
# настройка пределов обеих осей
plt.plot(x, np.sin(x), color="r", linestyle="solid")
plt.axis([-5, 5, -2, 2])
plt.show()
```

```python
# Заголовок и метки осей

plt.plot(x, np.sin(x-2), color="b", linestyle="dotted")
plt.title("Синусоида")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
# Легенда графика
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.legend()
plt.title("Синус и косинус")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Интерактивный режим
# %matplotlib notebook
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.legend()
plt.title("Синус и косинус")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

- Гистограммы

```python
# имитационные данные
data = np.random.randn(1000)
data
```

```python
# рисуем гистограмму
plt.hist(data, bins=50)
plt.show()
```

```python
# Несколько гистограмм в одном окне
# имитационные данные
x1 = np.random.normal(0, 0.5, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

# рисуем гистограммы
plt.hist(x1, bins=50, alpha=0.7, color="r", label="x1")
plt.hist(x2, bins=50, alpha=0.7, color="g", label="x2")
plt.hist(x3, bins=50, alpha=0.7, color="y", edgecolor="b", label="x3")
plt.legend()
plt.title("Гистограммы Гаусса")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
